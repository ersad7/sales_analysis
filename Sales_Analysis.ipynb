```python
df = pd.read_csv('sales_data.csv')

df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Drop the NaN values
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
# Removing the text in order date column
df = df[df['Order Date'].str[0:2]!='Or']
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
date = df['Order Date'].iloc[0]
date
```

```python
# Adding a Month Column
df['Month'] = df['Order Date'].apply(lambda time: time.month)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Changing the dtype of the columns
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df.dtypes
```

```python
# Adding City & State Column
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)}")
df['State'] = df['Purchase Address'].apply(lambda x: f"{get_state(x)}")

df.head()
```

```python
#Sales by Order Column
df['Sales'] = df['Quantity Ordered']*df['Price Each']

df.head()
```

#### Products most often sold together

```python
# Figure out the duplicate rows
d_df = df[df['Order ID'].duplicated(keep=False)]

# Dataframe is stripped down to only with duplicate values
d_df.head(10)
```

```python
# Grouping them together & Drop duplicates
d_df['Grouped'] = d_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
d_df = d_df[['Order ID','Grouped']].drop_duplicates()

d_df.head()
```

```python
# Counting Product Pairs
d_df.loc[:,'Grouped'].value_counts().head(10)
```

```python
# Counting Product Pairs
from itertools import combinations
from collections import Counter

count = Counter()
for row in d_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

```python
# Counting Product Pairs of 3
from itertools import combinations
from collections import Counter

count = Counter()
for row in d_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))
    
for key, value in count.most_common(10):
    print(key, value)
```

####  The most sold product

```python
product = df.groupby('Product').sum()['Quantity Ordered']
product.sort_values(ascending=False)
```

```python
product.plot(kind='bar')
```

```python
most_sold_products = pd.DataFrame(df.groupby(by='Product')['Quantity Ordered'].sum())
most_sold_products.reset_index(inplace=True, drop=False)
most_sold_products.columns = ['Product', 'Number of items sold']

sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(x=most_sold_products.iloc[:,0], y=most_sold_products.iloc[:,1])
plt.tight_layout()
```

```python
prices.plot(kind='line')
```

```python
# Mapping Prices over the Number of sales for a Product
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.set_ylabel('Quantity Ordered')
ax2.set_ylabel('Price ($)')

product.plot(kind='bar', ax=ax1,color='g')
prices.plot(kind='line', ax=ax2, color='r')
```
